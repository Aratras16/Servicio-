```python
def poligono(lados,rotacion = 0,show = '' ,esc = True,ql = 2):
    separacion = int(12/lados)
    poligono = stream.Stream()
    if rotacion >= 12:
        if esc == True:    
            for i in range(lados):
                lado = note.Note(escala_cromatica_rev[(i*separacion+rotacion) % 12],quarterlenght = ql)
                poligono.append(lado)
            if show == '':
                return poligono
            else:
                return poligono.show(show)
        else:
            for i in range(lados):
                lado = note.Note(circulo_Mayor[(i*separacion+rotacion) % 12])
                poligono.append(lado)
            return poligono.show(show)
        
    else:
        if esc == True:    
            for i in range(lados):
                lado = note.Note(escala_cromatica[(i*separacion+rotacion) % 12],quarterlenght = ql)
                poligono.append(lado)
            if show == '':
                return poligono
            else:
                return poligono.show(show)
        else:
            for i in range(lados):
                lado = note.Note(circulo_Mayor[(i*separacion+rotacion) % 12])
                poligono.append(lado)
            return poligono.show(show)
        
```

```python
def freq(lista):
    frecuencias = []
    for i in lista:
        a = pitch.Pitch(i)
        frecuencias.append(a.frequency)
    return frecuencias
```

```python
def midi(lista):
    midi = []
    for i in lista:
        a = pitch.Pitch(i)
        midi.append(a.midi)
    return midi
```

```python
def freq_med(lista):
    frecuencias_medias = []
    i = 0
    while i < int(len(lista))-1:
        medio = abs(float(lista[i+1])+float(lista[i]))/2
        frecuencias_medias.append(medio)
        i += 1
    return frecuencias_medias
```

```python
def midi_med(lista):
    midi_medio= []
    i = 0
    while i < int(len(lista))-1:
        medio = int(abs(int(lista[i+1])+int(lista[i]))/2)
        midi_medio.append(medio)
        i += 1
    return midi_medio
```
